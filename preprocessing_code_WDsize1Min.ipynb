```python
import pandas as pd
import numpy as np
import datetime
import re
```

```python
df = pd.read_csv("침착맨단독_chatlog.txt",sep=' ',names=['timeline','userId','comment'] ,encoding ="utf-8")
df
```

```python
timeline = df['timeline']
timeline
```

```python
df_temp = df.copy()
```

```python
df_temp["timeline"] = df_temp["timeline"].apply(lambda x: x.replace('[',''))
df_temp["timeline"] = df_temp["timeline"].apply(lambda x: x.replace(']',''))
df_temp
```

```python
df_temp["timeline"] = df_temp["timeline"].apply(lambda x: datetime.datetime.strptime(x, '%H:%M:%S'))
```

```python
df_temp
```

```python
startWindow = datetime.datetime.strptime("00:01:00",'%H:%M:%S')
classNum = 1
startWindow
```

```python
class_list = []
for i in range(df_temp.shape[0]):
    if (startWindow -  df_temp['timeline'][i]).total_seconds() > 0 :
        class_list.append(classNum)
    else:
        startWindow = startWindow + datetime.timedelta(minutes=1)
        classNum+=1
        class_list.append(classNum)
```

```python
df_temp['class'] = class_list
df_temp
```

```python
len(df_temp[df_temp['class'] == 1]['userId'])  #총 댓글 수
len(df_temp[df_temp['class'] == 1]['userId'].unique()) # 참여 유저 수
```

```python
df_temp["com_len"] = df_temp["comment"].apply(lambda x: len(x))
df_temp
```

```python
df_temp["com_initial"] = df_temp["comment"].apply(lambda x: "".join(re.compile('[ㄱ-ㅎ|ㅏ-ㅣ]+').findall(x)))
df_temp
```

```python
df_temp["initial_len"] = df_temp["com_initial"].apply(lambda x: len(x))
df_temp
```

```python
df_temp["com_special"] = df_temp["comment"].apply(lambda x: x.count('~')+x.count('!')+x.count('@')+x.count('#')+x.count('$')+x.count('%')+x.count('^')+x.count('&')+x.count('*')+x.count('(')+x.count(')')+x.count('<')+x.count('>')+x.count(',')+x.count('.')+x.count('?')+x.count('/'))
df_temp
```

```python
df_temp['com_has_init']= df_temp['com_initial'].apply(lambda x:1 if x != '' else 0)
df_temp.head(50)
```

```python
df_temp["com_k"] = df_temp["comment"].apply(lambda x: x.count('ㅋ'))
df_temp.tail(50)
```

```python
df_temp.drop('com_initial', axis=1, inplace = True)
df_temp
```

```python
df_temp.groupby('class',as_index=False)['com_len'].mean() #댓글 평균 길이
```

```python
df_preprocessed = df_temp.groupby('class',as_index=False)['comment'].count()#윈도우 내댓글 총 개수
df_preprocessed.columns = ['class', 'com_cnt']
df_preprocessed
```

```python
df_preprocessed["class"] = df_preprocessed["class"].apply(lambda x: (datetime.datetime.strptime("00:00:00", '%H:%M:%S') + datetime.timedelta(minutes=x-1)).strftime('%H:%M:%S'))
df_preprocessed
```

```python
df_preprocessed['com_len'] = df_temp.groupby('class',as_index=False)['com_len'].mean()['com_len'] 
df_preprocessed
```

```python
user_cnt_lst = []

for i in range(1, df_preprocessed.shape[0]+1):
    user_cnt_lst.append(len(df_temp[df_temp['class'] == i]['userId'].unique()))
    
df_preprocessed['user_cnt'] = user_cnt_lst

df_preprocessed
```

```python
df_preprocessed['k_cnt'] = df_temp.groupby('class',as_index=False)['com_k'].sum()['com_k'] 
df_preprocessed
```

```python
df_preprocessed['init_cnt'] = df_temp.groupby('class',as_index=False)['initial_len'].sum()['initial_len'] 
df_preprocessed
```

```python
df_preprocessed['special_cnt'] = df_temp.groupby('class',as_index=False)['com_special'].sum()['com_special'] 
df_preprocessed
```

```python
df_preprocessed['init_chat_ratio'] = df_temp.groupby('class',as_index=False)['com_has_init'].sum()['com_has_init']
df_preprocessed['init_chat_ratio'] = df_preprocessed['init_chat_ratio'] / df_preprocessed['com_cnt']
df_preprocessed
```

전처리 종료

```python
TIMESTAMP_FIELD = df_preprocessed["class"]
TIMESTAMP_FIELD
```

```python
df_preprocessed = df_preprocessed.drop('class',axis = 1)
```

```python
df_preprocessed.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("darkgrid")
from sklearn.ensemble import IsolationForest
```

```python
model=IsolationForest(n_estimators=100, max_samples='auto', n_jobs=-1,
                      max_features=7, contamination=0.01)
model.fit(df_preprocessed.to_numpy())
```

```python
score = model.decision_function(df_preprocessed.to_numpy())
df_result = df_preprocessed.copy()
df_result['scores']= score
df_result
```

```python
df_result['timeline'] = TIMESTAMP_FIELD
```

```python
df_candidate = df_result.sort_values(by = ['scores'], axis = 0, ascending=True).head(10)
df_candidate 
```

```python
standard = df_result.quantile(q = 0.25,interpolation='nearest')
standard
```

```python
com_cnt_std = standard[0]
user_cnt_std = standard[2]
k_cnt_std = standard[3]
```

```python
df_final = df_candidate[(df_candidate['com_cnt'] > com_cnt_std) & (df_candidate['k_cnt'] > k_cnt_std ) & (df_candidate['user_cnt'] > user_cnt_std )]
df_final
```

```python
final_timeline = df_final['timeline']
final_timeline
```
